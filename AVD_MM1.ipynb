Usei o gerador de n√∫meros aleat√≥rios G1 aleat√≥rimente. Diferentemente do trabalho 1, a divis√£o pelo m√≥dulo foi feito no retorno da pr√≥pria fun√ß√£o e o retorno √© apenas o array.

G1 √© um gerador de n√∫meros aleat√≥rios, que, segundo o teste chi-square, mostrou ter uma distribui√ß√£o uniforme entre [0,1] dos valores gerados.



```python
def G1(n):
    #Fixos
    p1 = 16807 #7‚Åµ
    p2 = 2147483647 # (2¬≥¬π - 1)

    anterior = 1030 #Seed.
    atual = 0 

    vet_rn = []
    
    for i in range(1,n+1,1):
        atual = (p1* anterior)%p2
        anterior = atual

        vet_rn.append(atual)
    
    return np.array(vet_rn)/p2 #p2 √© o m√≥dulo
```

Aqui s√£o geradas tr√™s filas: n1, n2 e n3, com respectivamente 10^3, 10^6 e 10^9 elementos.

```python
n1 = G1(10**3)
n2 = G1(10**6)
n3 = G1(10**9)
```
